```python
import pandas as pd
from datetime import datetime
```

```python
papers = pd.read_csv("../keywords_summary/science_direct/merged_keyword_results.csv")
papers.info()
```

```python
papers
```

```python
# Giả sử dataframe của bạn là 'papers'
# Nếu cột 'submitted' chưa phải định dạng datetime, chuyển đổi nó:
papers['submitted'] = pd.to_datetime(papers['submitted'], format='%Y-%m-%d')

# Định nghĩa ngày bắt đầu và ngày kết thúc
start_date = datetime(2023, 1, 1)
end_date = datetime(2025, 6, 16)  # Thời điểm hiện tại

# Lọc các bài báo có ngày submitted trong khoảng thời gian
filtered_papers = papers[(papers['submitted'] >= start_date) & (papers['submitted'] <= end_date)]
```

```python
filtered_papers.to_csv('science_direct/science_direct_datefiltered.csv', index=False)
```
